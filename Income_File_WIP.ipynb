```python
# Dependencies
import pandas as pd
```

```python
# Name of the CSV file
income_file =  "Resources/us_income_zipcode.csv"
```

```python
# The correct encoding must be used to read the CSV
all_income_df = pd.read_csv(income_file)
```

```python
#preview 
#all_income_df5 = all_income_df.head()
#print(all_income_df5)
```

```python
# Create a DataFrame from a list of dictionaries.
columns_to_keep = ['ZIP', 'Households', 'Households Median Income (Dollars)', 'Households Mean Income (Dollars)']
zip_income_df = all_income_df[columns_to_keep]
```

```python
#Preview
#zip_income_df5 = zip_income_df.head()
#print(zip_income_df5)
```

```python
# Drop all rows with missing information
zip_income_df = zip_income_df.dropna(how='any')
```

```python
# The Zip column is the wrong data type. It should be a string for Zip & Households & interger for Income.
zip_income_df.dtypes
```

```python
# Use df.astype() method to convert the datatype of the Zip & household columns to string
zip_income_df = zip_income_df.astype({"ZIP":str})

# Use df.astype() method to convert the datatype of the income columns to interger
columns_to_convert_int = ['Households Median Income (Dollars)', 'Households Mean Income (Dollars)']
zip_income_df[columns_to_convert_int] = zip_income_df[columns_to_convert_int].astype(int)
```

```python
# confirm changes to string & interger.
zip_income_df.dtypes
```

```python
# We can calculate individual aggregate functions for Household column
print('Households:')
print(f"The count is {zip_income_df['Households'].count()}")
print(f"The minimum is {zip_income_df['Households'].min()}")
print(f"The maximum is {zip_income_df['Households'].max()}")
```

```python
# We can calculate individual aggregate functions for Households Median Income (Dollars) column
print('Households Median Income (Dollars)')
print(f"The count is {zip_income_df['Households Median Income (Dollars)'].count()}")
print(f"The minimum is {zip_income_df['Households Median Income (Dollars)'].min()}")
print(f"The maximum is {zip_income_df['Households Median Income (Dollars)'].max()}")
```

```python
# We can calculate individual aggregate functions for ouseholds Mean Income (Dollars) column
print('Households:')
print(f"The count is {zip_income_df['Households Mean Income (Dollars)'].count()}")
print(f"The minimum is {zip_income_df['Households Mean Income (Dollars)'].min()}")
print(f"The maximum is {zip_income_df['Households Mean Income (Dollars)'].max()}")
```

```python
##Sorting data so that highest median income shows at top
sorted_median_income_df = zip_income_df.sort_values(by='Households Median Income (Dollars)', ascending=True)
sorted_median_income_df.head(20)
```

```python
#Sorting data so that highest mean income shows at top
sorted_mean_income_df = zip_income_df.sort_values(by='Households Mean Income (Dollars)', ascending=True)
sorted_mean_income_df.head(20)
```

```python
#Sorting data so only "Upper income" household Median shows. Upper income is  $149,132
threshold = 149132
upper_median_df = zip_income_df[zip_income_df['Households Median Income (Dollars)'] > threshold]
upper_median_df.head(20)
```

```python
#Sorting data so only "Upper income" household Mean shows. Upper income is  $149,132
threshold = 149132
upper_mean_df = zip_income_df[zip_income_df['Households Mean Income (Dollars)'] > threshold]
upper_mean_df.head(20)
```

```python
# Save the DataFrame to a CSV file. 
zip_income_df.to_csv("Resources/zip_income_df.csv", index=False)
```
